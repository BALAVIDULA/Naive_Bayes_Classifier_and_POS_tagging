```python
rivewdata = rivewdata.drop(['User_ID', 'Browser_Used', 'Device_Used'], axis=1)
```

```python
count=rivewdata.isnull().sum().sort_values(ascending=False)
percentage=((rivewdata.isnull().sum()/len(rivewdata)*100)).sort_values(ascending=False)
missing_data=pd.concat([count,percentage],axis=1,keys=['Count','Percentage'])

missing_data
```

```python


import re
import string
def text_clean(text):
    text=text.lower()
    text=re.sub('\[.*?\]','',text)
    text=re.sub('[%s]'% re.escape(string.punctuation),'',text)
    rext=re.sub('\w*\d\w*','',text)
    return text
cleaned1=lambda x:text_clean(x)

```

```python
rivewdata['cleaned_description']=pd.DataFrame(rivewdata.Description.apply(cleaned1))
rivewdata.head(10)
```

```python
def text_clean_2(text):
    text=re.sub('[''""...]','',text)
    text=re.sub('\n','',text)
    return text
cleaned2=lambda x:text_clean_2(x)
```

```python


rivewdata['cleaned_description_new'] = pd.DataFrame(rivewdata['cleaned_description'].apply(cleaned2))
rivewdata.head(10)

```

```python
from sklearn.model_selection import train_test_split
Id=rivewdata.cleaned_description_new
de=rivewdata.Is_Response
i_train,i_test,d_train,d_test=train_test_split(Id,de,test_size=0.1,random_state=225)
print("i_train",len(i_train))
print("i_test",len(i_test))
print("d_train",len(d_train))
print("d_test",len(d_test))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
vectorizer = CountVectorizer()
i_train_vectorized = vectorizer.fit_transform(i_train)
i_test_vectorized = vectorizer.transform(i_test)
```

```python

classifier = MultinomialNB()
classifier.fit(i_train_vectorized, d_train)
```

```python

predictions = classifier.predict(i_test_vectorized)

accuracy = metrics.accuracy_score(d_test, predictions)
print("Accuracy:", accuracy)
```

```python

new_review = ["This hotel is fantastic!"]
new_review_vectorized = vectorizer.transform(new_review)
prediction = classifier.predict(new_review_vectorized)
print("Predicted Label:", prediction[0])
```
